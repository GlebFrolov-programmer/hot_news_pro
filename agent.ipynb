````python
# Нормализация кодировки в названии файлов. Проблема с кириллицей

import os
import unicodedata
import glob
from pathlib import Path

# Формируем путь к директории
directory = Path(mr_conf.OUTPUT_DIR_POST_PROCESSING)

# Найти все xlsx файлы в указанной директории
for filepath in glob.glob(str(directory / '*.json')):
    # print(filepath)
    # Нормализовать имя файла
    normalized_name = unicodedata.normalize('NFC', os.path.basename(filepath))

    if os.path.basename(filepath) != normalized_name:
        # Формируем новый полный путь
        new_filepath = os.path.join(directory, normalized_name)

        # Переименовываем файл
        os.rename(filepath, new_filepath)
        print(f'Исправлено: {filepath} -> {new_filepath}')
print('Готово')
````

````python
import operator
from typing_extensions import TypedDict
from langgraph.graph.message import add_messages

class EventResult(BaseModel):
    """Результат анализа события"""
    who: str = Field(description="Кто является участником/инициатором события")
    what: str = Field(description="Что именно произошло")
    where: str = Field(description="Где произошло событие")
    when: str = Field(description="Когда произошло/произойдет/планируется событие")
    why: str = Field(description="Почему это произошло")
    how: str = Field(description="Как это произошло")
    effect: str = Field(description="Какие есть последствия у события?")
    confidence: int = Field(description="Уверенность в точности информации (оценка от 1 до 10)", ge=1, le=10)
    event: str = Field(description="Итоговое новостное событие summary на основе вопросов")

class EventScore(BaseModel):
    """Структура для аналитической оценки события"""

    impact_scale: int = Field(
        description="Сила и масштаб влияния: значимость последствий для региона, количество затронутых сфер (экономика, соц. сфера, экология и т.д.) (1-10)", ge=1, le=10
    )

    regional_specificity: int = Field(
        description="Региональная специфичность: насколько событие характерно именно для данного региона, отражает ли его уникальные проблемы/особенности (1-10)",  ge=1, le=10
    )

    data_richness: int = Field(
        description="Богатство данных: событие содержит ответы на простые вопросы (кто, что, когда, где и тп) (1-10)", ge=1, le=10
    )

    correct_period: int = Field(
        description="Конкретный период: в событии упоминается конкретный период (1-10)", ge=1, le=10
    )

    statistic_data: int = Field(
        description="Статистические данные: в событии приводится большое количество цифр (не включаю дат и периодов) (1-10)", ge=1, le=10
    )

    no_promo_info: int = Field(
        description="Реклама: само событие представляет просто рекламу какого-то события или продукта/объекта (1-10)", ge=1, le=10
    )

    trusted_source: int = Field(
        description="Проверенные источники: упонимается в событии статистика от Сбер, ДомКлик, Сбериндекс, Росстат (1-10)", ge=1, le=10
    )

    total_score: float = Field(description="Итоговая оценка", ge=1, le=10)

    analytics_notes: str = Field(default="", description="Комментарий аналитика: краткое обоснование выставленных оценок, ключевые моменты"
    )

class ScoredEvent(BaseModel):
    """Событие с оценкой"""
    event: EventResult
    score: EventScore

class UniqueEventWithFrequency(BaseModel):
    """Уникальное событие с частотой встречаемости"""
    event: EventResult
    score: EventScore
    frequency: int = Field(default=1, description="Количество раз, которое это событие встречалось")
    source_records: List[str] = Field(default_factory=list, description="ID исходных записей, содержащих это событие")

# LangGraph State
class AgentState(TypedDict):
    """Состояние агента для LangGraph"""
    messages: Annotated[list, add_messages]
    text: str
    region: str
    category: str
    period: str
    current_event: Optional[Dict[str, Any]]
    current_score: Optional[Dict[str, Any]]
    is_unique: bool
    duplicate_index: int
    action: str
    processing_status: str
    error: Optional[str]
    raw_source: Optional[Dict[str, Any]]

class GlobalEventsStore:
    """Глобальное хранилище для всех событий между вызовами агента"""

    _instance = None
    _events: List[Dict] = []
    _next_id: int = 0  # Счетчик для генерации ID

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super(GlobalEventsStore, cls).__new__(cls)
        return cls._instance

    def _generate_id(self) -> int:
        """Генерирует уникальный ID для события"""
        event_id = self._next_id
        self._next_id += 1
        return event_id

    def add_event(self, event_data: Dict, frequency: int = 1, raw_source: Dict = None) -> int:
        """Добавляет событие в хранилище и возвращает его ID"""
        event_id = self._generate_id()

        event_with_freq = {
            "id": event_id,  # Добавляем ID события
            **event_data,
            "frequency": frequency,
            "added_at": datetime.now().isoformat(),
            "source_records": [event_data.get('source_record_id', 'unknown')],
            "raw_source": [raw_source] if raw_source else []  # Массив с исходным контентом
        }
        self._events.append(event_with_freq)
        self.print_stats_frequency()
        print(f"✅ Событие {event_id} добавлено в глобальное хранилище. Всего событий: {len(self._events)}")
        return event_id

    def update_event_frequency(self, event_id: int, increment: int = 1, source_record_id: str = None, raw_source: Dict = None) -> bool:
        """Увеличивает частоту существующего события по ID и добавляет source_record и raw_source"""
        print(f"🔍 Ищу событие с ID {event_id} для обновления frequency...")
        for event in self._events:
            if event.get('id') == event_id:
                old_frequency = event.get("frequency", 1)
                event["frequency"] += increment

                # Добавляем source_record если передан
                if source_record_id:
                    if "source_records" not in event:
                        event["source_records"] = []
                    event["source_records"].append(source_record_id)

                # Добавляем raw_source в массив если передан
                if raw_source:
                    event["raw_source"].append(raw_source)
                    print(f"   📋 Добавлен raw_source в массив. Всего: {len(event['raw_source'])}")

                print(f"📈 Обновлена частота события {event_id}: {old_frequency} -> {event['frequency']}")
                return True

        print(f"❌ Событие с ID {event_id} не найдено в хранилище")
        print(f"   Доступные ID: {[e.get('id') for e in self._events]}")
        return False

    def get_events(self) -> List[Dict]:
        """Возвращает все события"""
        return self._events.copy()

    def get_event_by_id(self, event_id: int) -> Optional[Dict]:
        """Возвращает событие по ID"""
        for event in self._events:
            if event.get('id') == event_id:
                return event
        return None

    def get_event_by_index(self, index: int) -> Optional[Dict]:
        """Возвращает событие по индексу в списке"""
        if 0 <= index < len(self._events):
            return self._events[index]
        return None

    def clear_events(self) -> None:
        """Очищает хранилище (для новой задачи)"""
        events_count = len(self._events)
        self._events.clear()
        self._next_id = 0  # Сбрасываем счетчик ID
        print(f"🗑️ Глобальное хранилище событий очищено (удалено {events_count} событий)")

    def print_stats_frequency(self) -> None:
        print([f'id({event.get("id", "unknown")}): {event.get("frequency", "???")}' for event in self._events])

    def get_stats(self) -> Dict[str, Any]:
        """Возвращает статистику по событиям"""
        total_frequency = sum(event.get('frequency', 1) for event in self._events)
        total_raw_sources = sum(len(event.get('raw_source', [])) for event in self._events)

        return {
            "total_unique_events": len(self._events),
            "total_occurrences": total_frequency,
            "total_raw_sources": total_raw_sources,
            "avg_frequency": total_frequency / len(self._events) if self._events else 0,
            "avg_raw_sources_per_event": total_raw_sources / len(self._events) if self._events else 0
        }

def get_global_events_store() -> GlobalEventsStore:
    """Возвращает глобальный экземпляр хранилища событий"""
    return GlobalEventsStore()
````

````python
from langgraph.graph import StateGraph, END

class EventProcessingGraph:
    """LangGraph реализация агента обработки событий"""

    def __init__(self, model: str = "GigaChat-2-Max"):
        print("🤖 ИНИЦИАЛИЗАЦИЯ LANGGRAPH АГЕНТА...")
        self.model = model
        # self.llm = GigaChat(
        #     base_url='https://gigachat-ift.sberdevices.delta.sbrf.ru/v1',
        #     cert_file='giga.pem',
        #     key_file='giga.key',
        #     verify_ssl_certs=False,
        #     model=model,
        #     verbose=False,
        #     temperature=0.1,
        #     timeout=300
        # )
        self.llm = ChatOpenAI(
                base_url=os.getenv('OPENAI_API_URL'),
                api_key=os.getenv('OPENAI_API_KEY'),
                model=os.getenv('OPENAI_MODEL_NAME'),
                verbose=False,
                temperature=0.1,
                timeout=300
            )

        # Создаем граф
        self.graph = self._create_graph()
        print("✅ LANGGRAPH АГЕНТ УСПЕШНО ИНИЦИАЛИЗИРОВАН")

    def _create_graph(self) -> StateGraph:
        """Создает граф обработки событий"""
        workflow = StateGraph(AgentState)

        # Добавляем узлы
        workflow.add_node("extract_event", self._extract_event_node)
        workflow.add_node("check_uniqueness", self._check_uniqueness_node)
        workflow.add_node("score_event", self._score_event_node)
        # workflow.add_node("handle_empty_add_messagesevent", self._handle_empty_event_node)
        workflow.add_node("handle_empty_event", self._handle_empty_event_node)
        workflow.add_node("finalize_result", self._finalize_result_node)

        # Устанавливаем начальную точку
        workflow.set_entry_point("extract_event")

        # Добавляем ребра
        workflow.add_conditional_edges(
            "extract_event",
            self._should_check_uniqueness,
            {
                "check_uniqueness": "check_uniqueness",
                "empty_event": "handle_empty_event"
            }
        )

        workflow.add_conditional_edges(
            "check_uniqueness",
            self._should_score_event,
            {
                "score_event": "score_event",
                "skip_scoring": "finalize_result"
            }
        )

        workflow.add_edge("score_event", "finalize_result")
        workflow.add_edge("handle_empty_event", "finalize_result")
        workflow.add_edge("finalize_result", END)

        return workflow.compile()

    def _extract_event_node(self, state: AgentState) -> AgentState:
        """Узел извлечения события"""
        print(f"\n🎯 УЗЕЛ: ИЗВЛЕЧЕНИЕ СОБЫТИЯ")

        tool = EventExtractorTool()
        event_json = tool._run(state["text"], state["region"], state["category"], state["period"])

        try:
            event_data = json.loads(event_json) if event_json else {}
            state["current_event"] = event_data
            print(f"✅ Событие извлечено: {bool(event_data)}")
        except Exception as e:
            print(f"❌ Ошибка парсинга события: {e}")
            state["current_event"] = {}
            state["error"] = f"Ошибка извлечения: {e}"

        return state

    def _check_uniqueness_node(self, state: AgentState) -> AgentState:
        """Узел проверки уникальности"""
        print(f"\n🔍 УЗЕЛ: ПРОВЕРКА УНИКАЛЬНОСТИ")

        if not state["current_event"]:
            state["is_unique"] = True
            state["duplicate_index"] = -1
            return state

        tool = EventUniquenessCheckerTool()
        event_json = json.dumps(state["current_event"], ensure_ascii=False)
        uniqueness_result = tool._run(event_json)

        try:
            uniqueness_data = json.loads(uniqueness_result)
            state["is_unique"] = uniqueness_data.get("is_unique", True)
            state["duplicate_index"] = uniqueness_data.get("duplicate_index", -1)
            print(f"✅ Уникальность проверена: {state['is_unique']}")
        except Exception as e:
            print(f"❌ Ошибка проверки уникальности: {e}")
            state["is_unique"] = True
            state["duplicate_index"] = -1

        return state

    def _score_event_node(self, state: AgentState) -> AgentState:
        """Узел оценки события"""
        print(f"\n📊 УЗЕЛ: ОЦЕНКА СОБЫТИЯ")

        if not state["current_event"] or not state["is_unique"]:
            state["current_score"] = {}
            return state

        tool = EventScorerTool()
        event_json = json.dumps(state["current_event"], ensure_ascii=False)
        score_json = tool._run(event_json, state["region"], state["category"], state["period"])

        try:
            score_data = json.loads(score_json)
            state["current_score"] = score_data
            print(f"✅ Событие оценено: {score_data.get('total_score', 0)}/10")
        except Exception as e:
            print(f"❌ Ошибка оценки: {e}")
            state["current_score"] = {}
            state["error"] = f"Ошибка оценки: {e}"

        return state

    def _handle_empty_event_node(self, state: AgentState) -> AgentState:
        """Узел обработки пустого события"""
        print(f"\n🔄 УЗЕЛ: ОБРАБОТКА ПУСТОГО СОБЫТИЯ")

        state["current_event"] = {}
        state["current_score"] = {}
        state["is_unique"] = False
        state["duplicate_index"] = -1
        state["action"] = "not_found"

        return state

    def _finalize_result_node(self, state: AgentState) -> AgentState:
        """Финальный узел - подготовка результата"""
        print(f"\n🎉 УЗЕЛ: ФИНАЛИЗАЦИЯ РЕЗУЛЬТАТА")

        # Определяем действие
        if not state["current_event"]:
            state["action"] = "not_found"
            state["processing_status"] = "no_event"
        elif not state["is_unique"]:
            state["action"] = "updated"
            state["processing_status"] = "duplicate"
        else:
            state["action"] = "added"
            state["processing_status"] = "success"

        print(f"✅ Результат финализирован: {state['action']}")
        return state

    def _should_check_uniqueness(self, state: AgentState) -> str:
        """Условное ребро: проверять уникальность или обработать как пустое"""
        if state["current_event"]:
            return "check_uniqueness"
        return "empty_event"

    def _should_score_event(self, state: AgentState) -> str:
        """Условное ребро: оценивать событие или пропустить оценку"""
        if state["current_event"] and state["is_unique"]:
            return "score_event"
        return "skip_scoring"

    def process_text(self, text: str, region: str, category: str, period: str, raw_source: Dict = None) -> Dict[str, Any]:
        """Основной метод обработки текста"""
        print(f"\n{'='*70}")
        print("🚀 НАЧАЛО РАБОТЫ LANGGRAPH АГЕНТА")
        print(f"📝 Текст: {text[:150]}...")
        print(f"📍 Регион: {region}")
        print(f"📊 Категория: {category}")
        print(f"📊 Период: {period}")
        print(f"{'='*70}")

        try:
            # Инициализируем состояние
            initial_state: AgentState = {
                "messages": [],
                "text": text,
                "region": region,
                "category": category,
                "period": period,
                "current_event": None,
                "current_score": None,
                "is_unique": True,
                "duplicate_index": -1,
                "action": "unknown",
                "processing_status": "processing",
                "error": None,
                "raw_source": raw_source
            }

            start_time = time.time()
            # Выполняем граф
            final_state = self.graph.invoke(initial_state)
            elapsed = time.time() - start_time

            print(f"\n✅ ГРАФ ВЫПОЛНЕН за {elapsed:.1f}с")
            print(f"📊 РЕЗУЛЬТАТ: {final_state['action']}")

            return self._prepare_result(final_state)

        except Exception as e:
            print(f"❌ ОШИБКА В ГРАФЕ: {e}")
            import traceback
            traceback.print_exc()
            return self._create_error_result(text, region, category, str(e))

    def _prepare_result(self, state: AgentState) -> Dict[str, Any]:
        """Подготавливает финальный результат"""
        result = {
            "event": state.get("current_event", {}),
            "score": state.get("current_score", {}),
            "is_unique": state.get("is_unique", True),
            "duplicate_index": state.get("duplicate_index", -1),
            "action": state.get("action", "unknown"),
            "processing_status": state.get("processing_status", "error"),
            "error": state.get("error")
        }

        # Создаем Pydantic модели если есть данные
        try:
            if result["event"] and result["score"]:
                event_model = EventResult(**result["event"])
                score_model = EventScore(**result["score"])
                scored_event = ScoredEvent(event=event_model, score=score_model)
                result["scored_event"] = scored_event
        except Exception as e:
            print(f"⚠️ Ошибка создания Pydantic моделей: {e}")
            result["scored_event"] = None

        return result

    def _create_error_result(self, text: str, region: str, category: str, error: str) -> Dict[str, Any]:
        """Создает результат при ошибке"""
        return {
            "event": {},
            "score": {},
            "is_unique": False,
            "duplicate_index": -1,
            "action": "error",
            "processing_status": "error",
            "error": error,
            "scored_event": None
        }
````

````python
from langchain_core.messages import SystemMessage, HumanMessage

output_dir_post_processing = mr_conf.OUTPUT_DIR_POST_PROCESSING
output_dir_events = mr_conf.OUTPUT_DIR_EVENTS  # Новая папка для событий
retry_blacklist = 1

# Инициализируем LangGraph агента
print("🤖 ИНИЦИАЛИЗАЦИЯ LANGGRAPH АГЕНТА ДЛЯ ГЕНЕРАЦИИ СОБЫТИЙ...")
event_agent = EventProcessingGraph()
print("✅ LANGGRAPH АГЕНТ УСПЕШНО ИНИЦИАЛИЗИРОВАН")

# Класс для обработки сырых данных с помощью GigaChat
class RawDataProcessor:
    def __init__(self, model="GigaChat"):
        # self.llm = GigaChat(
        #     base_url='https://gigachat-ift.sberdevices.delta.sbrf.ru/v1',
        #     cert_file='giga.pem',
        #     key_file='giga.key',
        #     verify_ssl_certs=False,
        #     model=model,
        #     verbose=False,
        #     temperature=0.1,
        #     timeout=300
        # )
        self.llm = ChatOpenAI(
                base_url=os.getenv('OPENAI_API_URL'),
                api_key=os.getenv('OPENAI_API_KEY'),
                model=os.getenv('OPENAI_MODEL_NAME'),
                verbose=False,
                temperature=0.1,
                timeout=300
            )

    def process_event_raw_data(self, event, raw_contents):
        """
        Обрабатывает сырые данные события через GigaChat для создания short_raw_data

        Args:
            event: словарь с данными события
            raw_contents: список сырых контентов (максимум 3)

        Returns:
            str: очищенный и детализированный текст short_raw_data
        """
        if not raw_contents:
            return ""

        try:
            # Формируем промт для обработки сырых данных С УЧЕТОМ ИНФОРМАЦИИ О СОБЫТИИ
            prompt = HumanMessage(self._build_prompt(event, raw_contents))

            # Отправляем запрос к модели
            response = self.llm.invoke([prompt])
            processed_text = response.content.strip()

            return processed_text

        except Exception as e:
            print(f"❌ Ошибка обработки сырых данных для события: {e}")
            # В случае ошибки возвращаем объединенный текст без обработки
            return "\n".join([content[:500] for content in raw_contents if content])

    def _build_prompt(self, event, raw_contents):
        """Строит промт для обработки сырых данных с учетом информации о событии"""

        # Извлекаем ключевую информацию о событии
        event_data = event.get('event', {})
        event_summary = event_data.get('event', '')  # основное описание события
        who = event_data.get('who', '')
        what = event_data.get('what', '')
        when = event_data.get('when', '')
        where = event_data.get('where', '')

        contents_text = "\n\n".join([
            f"Исходный текст {i+1}:\n{content}"
            for i, content in enumerate(raw_contents)
        ])

        prompt = f"""
Ты — профессиональный редактор и аналитик. Тебе нужно обработать несколько текстов, описывающих одно и то же событие.

ИНФОРМАЦИЯ О СОБЫТИИ (уже выделенная суть):
- Основное описание: {event_summary}
- Кто участвует: {who}
- Что произошло: {what}
- Когда: {when}
- Где: {where}

Исходные тексты для обработки:

{contents_text}

Задача:
1. Используя информацию о событии как ориентир, выдели из сырых текстов ВСЕ детали, относящиеся к этому событию
2. Убрать весь мусор (рекламные вставки, не относящуюся информацию, повторы)
3. Удалить дублирование информации, но при этом дополнить друг друга, если в разных текстах есть разные детали
4. Сохранить максимальную детализацию и суть события, особенно то, что относится к указанному выше событию
5. В результате должен получиться один связный, очищенный текст, который содержит всю важную информацию из представленных текстов, относящуюся к данному событию
6. Длина итогового текста должна получится не более 5 предложений.

Требования к результату:
- Не добавляй спецсимволы и лишиние водные слова. Только текст.
- Фокусироваться на информации, относящейся к указанному событию
- Без сжатия сути и ключевых деталей
- Хронологическая последовательность, если это уместно
- Конкретные факты, цифры, имена, даты
- Причины и последствия события, если они упоминаются
- Детали, которые делают описание полным и информативным
- Если в текстах есть противоречивая информация - сохранить основные факты

Обработанный текст (должен быть готов к публикации, без пометок о процессе обработки):
"""
        return prompt

# Инициализируем процессор сырых данных
raw_data_processor = RawDataProcessor()

def calculate_and_sort_events(events_to_save):
    """
    Рассчитывает event_score для каждого события и сортирует массив по убыванию этого значения.

    Формула: total_score * K + (frequency/max(frequency)) * (1 - K) * 10

    Args:
        events_to_save: массив словарей с событиями

    Returns:
        Отсортированный массив событий с рассчитанным event_score
    """
    if not events_to_save:
        return events_to_save

    # Находим максимальную частоту
    # max_freq = max(event.get('frequency', 1) for event in events_to_save)
    weight_approved_source = 1
    max_freq = max(
        event.get('frequency', 1) + sum(weight_approved_source for source in event.get('raw_source', []) if source.get('approved', 0) is True)
        for event in events_to_save
    )

    # Рассчитываем event_score для каждого события
    for event in events_to_save:
        total_score = event.get('score', {}).get('total_score', 1)
        # frequency = event.get('frequency', 1)
        frequency = event.get('frequency', 1) + sum(weight_approved_source for source in event.get('raw_source', []) if source.get('approved', 0) is True)
        # Расчет по формуле
        normalized_frequency = frequency / max_freq * 10 if max_freq > 0 else 0
        k = 0.65
        event_score = total_score * k + normalized_frequency * (1 - k)

        # Сохраняем рассчитанное значение
        event['frequency'] = frequency
        event['event_score_calculation'] = {
            'Доля влияния total_score': k,
            'Доля влияния frequency': 1-k,
            'max_freq': max_freq,
            'normalized_frequency': normalized_frequency,
            'total_score': total_score,
            'calc_normalized_frequency': normalized_frequency * (1 - k),
            'calc_total_score': total_score * k
        }
        event['event_score'] = event_score

    # Сортируем по event_score по убыванию
    sorted_events = sorted(
        events_to_save,
        key=lambda x: x.get('event_score', 0),
        reverse=True
    )

    return sorted_events

def process_raw_data_with_agent(raw_data, region, category, period, raw_source=None):
    """Обрабатывает сырые данные через LangGraph агента для генерации событий"""
    normalized_data = unicodedata.normalize('NFC', raw_data)
    if not normalized_data or len(normalized_data.strip()) < 10:
        return None, False

    try:
        # Используем LangGraph агента для генерации события
        result = event_agent.process_text(
            text=normalized_data,
            region=region,
            category=category,
            period=period,
            raw_source=raw_source
        )
        return result, False
    except Exception as e:
        print(f'ERROR GENERATE EVENT: {e}')
        return None, False

def is_empty_event_result(result: Dict) -> bool:
    """Проверяет, является ли результат пустым"""
    if not result:
        return True

    event_data = result.get('event', {})
    if not event_data:
        return True

    # Проверяем по содержимому полей события
    who = event_data.get('who', '').strip()
    what = event_data.get('what', '').strip()
    event_text = event_data.get('event', '').strip()
    confidence = event_data.get('confidence', 1)

    if (not who and not what and not event_text) or confidence < 2:
        return True

    return False

# Кастомный JSON энкодер для обработки Pydantic моделей и других объектов
class PydanticEncoder(json.JSONEncoder):
    def default(self, obj):
        if hasattr(obj, 'dict'):
            return obj.dict()
        elif isinstance(obj, datetime):
            return obj.isoformat()
        elif hasattr(obj, '__dict__'):
            return str(obj)
        return super().default(obj)

def convert_to_serializable(obj):
    """Рекурсивно преобразует объект в сериализуемые данные"""
    if obj is None:
        return None
    elif hasattr(obj, 'dict'):
        return obj.dict()
    elif isinstance(obj, (str, int, float, bool)):
        return obj
    elif isinstance(obj, datetime):
        return obj.isoformat()
    elif isinstance(obj, dict):
        return {k: convert_to_serializable(v) for k, v in obj.items()}
    elif isinstance(obj, list):
        return [convert_to_serializable(item) for item in obj]
    elif hasattr(obj, '__dict__'):
        return str(obj)
    else:
        return str(obj)

def process_top_events_with_raw_data(events_to_save, top_n=5):
    """
    Обрабатывает топ-N событий, добавляя поле short_raw_data

    Args:
        events_to_save: список событий
        top_n: количество топ-событий для обработки

    Returns:
        Список событий с добавленным полем short_raw_data для топ-N
    """
    if not events_to_save:
        return events_to_save

    print(f"🔍 ОБРАБОТКА ТОП-{top_n} СОБЫТИЙ ДЛЯ ФОРМИРОВАНИЯ SHORT_RAW_DATA...")

    top_events = events_to_save[:top_n]

    for i, event in enumerate(top_events):
        print(f"   Обрабатываю событие {i+1}/{len(top_events)} (ID: {event.get('id')})")

        raw_sources = event.get('raw_source', [])

        if not raw_sources:
            print(f"   ⚠️ Нет сырых источников для события {event.get('id')}")
            event['short_raw_data'] = ""
            continue

        # Сортируем сырые источники: approved=True в начало
        sorted_sources = sorted(
            raw_sources,
            key=lambda x: x.get('approved', False),
            reverse=True  # True будет первыми
        )

        # Берем первые 3 источника
        selected_sources = sorted_sources[:3]

        # Извлекаем сырые данные
        raw_contents = []
        for j, source in enumerate(selected_sources):
            raw_data = source.get('raw_data', '')
            if raw_data and isinstance(raw_data, str) and len(raw_data.strip()) > 0:
                raw_contents.append(raw_data.strip())
                print(f"      📄 Источник {j+1}: approved={source.get('approved', False)}, длина={len(raw_data)}")

        if not raw_contents:
            print(f"   ⚠️ Нет валидных сырых данных для события {event.get('id')}")
            event['short_raw_data'] = ""
            continue

        print(f"   🔄 Обрабатываю {len(raw_contents)} сырых контентов через GigaChat...")

        # Обрабатываем через GigaChat
        try:
            short_raw_data = raw_data_processor.process_event_raw_data(event, raw_contents)
            event['short_raw_data'] = short_raw_data
            print(f"   ✅ Успешно обработано, длина результата: {len(short_raw_data)}")
        except Exception as e:
            print(f"   ❌ Ошибка обработки через GigaChat: {e}")
            # В случае ошибки объединяем первые части сырых данных
            fallback_text = "\n".join([content[:500] for content in raw_contents])
            event['short_raw_data'] = fallback_text

    return events_to_save

for task in tqdm(tasks_to_parse):
    start_time = time.time()
    print(f'''
    **************** TASK ({tasks_to_parse.index(task)+1}/{len(tasks_to_parse)}) {task.metadata} ****************
    ''')

    # ОЧИСТКА ХРАНИЛИЩА ДЛЯ НОВОЙ ЗАДАЧИ
    event_store = get_global_events_store()
    event_store.clear_events()
    print("🔄 Хранилище событий очищено для новой задачи")

    # Получаем регион и категорию из метаданных задачи
    region = task.metadata.get('AVAILABLE_REGIONS', 'Неизвестный регион')
    category = task.metadata.get('AVAILABLE_CATEGORIES', 'Неизвестная категория')
    period = task.metadata.get('PERIOD', 'Неизвестный период')

    print(f"📍 Регион: {region}")
    print(f"📊 Категория: {category}")

    # Проверяем, существует ли файл событий
    output_filename = f"EVENTS_{task.parameters.get('TEMPLATES_FILENAME_BASE').format(**task.metadata)}.json"
    output_path = os.path.join(output_dir_events, output_filename)

    if os.path.exists(output_path):
        print(f"✅ Файл {output_path} уже существует. Пропускаем задачу.")
        continue

    filename = os.path.join(
        output_dir_post_processing,
        f"POST_PROCESSING_{task.parameters.get('TEMPLATES_FILENAME_BASE').format(**task.metadata)}.json"
    )

    try:
        # Чтение исходного json массива словарей
        with open(filename, 'r', encoding='utf-8') as f:
            records = json.load(f)
        print(f"📁 Загружено {len(records)} записей для обработки")
    except Exception as e:
        print(f'❌ ОШИБКА ЧТЕНИЯ ФАЙЛА {filename}: {e}')
        continue

    # Обрабатываем каждую запись через LangGraph агента
    successful_events = 0
    failed_events = 0
    duplicate_events = 0
    empty_events = 0

    c = 0
    for record in tqdm(records, desc="Обработка записей"):
        raw = record.get('raw_data')
        # try:
        #     open(os.path.join(mr_conf.OUTPUT_DIR_PATH, 'logs', str(c) + ') ' + __import__('datetime').datetime.now().strftime('%Y-%m-%d_%H-%M-%S')+'.txt'), 'w', encoding='utf-8').write(raw[:10])
        # except Exception as e:
        #     print(e)
        # c += 1
        if not raw or len(raw.strip()) < 10:
            record['scored_event'] = None
            record['processing_status'] = 'skipped_empty'
            failed_events += 1
            continue

        try:
            # Генерируем событие через LangGraph агента
            result, is_blacklist = process_raw_data_with_agent(raw, region, category, period, record)

            if result and not is_empty_event_result(result):
                processing_status = result.get('processing_status')
                is_unique = result.get('is_unique', True)
                duplicate_index = result.get('duplicate_index', -1)

                # ПОДГОТОВКА ДАННЫХ ДОЛЖНА БЫТЬ ВНЕ УСЛОВИЙ
                event_data = convert_to_serializable(result.get('event', {}))
                score_data = convert_to_serializable(result.get('score', {}))

                full_event_data = {
                    "event": event_data,
                    "score": score_data,
                    "region": region,
                    "category": category,
                    "period": period
                }

                print(f"🔍 Анализ результата: статус={processing_status}, уникальное={is_unique}, дубликат_индекс={duplicate_index}")

                # ИСПРАВЛЕННОЕ УСЛОВИЕ - используем processing_status вместо is_unique
                if processing_status == 'success':
                    # Уникальное событие - добавляем в хранилище
                    event_store = get_global_events_store()

                    # Сохраняем raw_source и возвращаем ID события
                    event_id = event_store.add_event(full_event_data, frequency=1, raw_source=record)
                    print(f"🎉 Уникальное событие {event_id} добавлено в глобальное хранилище")

                    # Обновляем запись сериализуемыми данными
                    serializable_result = convert_to_serializable(result)
                    record.update(serializable_result)
                    record['event_id'] = event_id  # Сохраняем ID события
                    successful_events += 1

                elif processing_status == 'duplicate':
                    # Дубликат - увеличиваем частоту существующего события
                    if duplicate_index >= 0:
                        # Получаем событие по индексу
                        duplicate_event = event_store.get_event_by_index(duplicate_index)
                        if duplicate_event:
                            event_id = duplicate_event.get('id')
                            print(f"🔍 Найден дубликат: ID={event_id}, текущая частота={duplicate_event.get('frequency', 1)}")

                            success = event_store.update_event_frequency(
                                event_id,
                                increment=1,
                                source_record_id=record.get('id', 'unknown'),
                                raw_source=record
                            )

                            if success:
                                print(f"🔁 Дубликат обнаружен, увеличена частота события {event_id}")

                                # Сохраняем информацию о дубликате в запись
                                serializable_result = convert_to_serializable(result)
                                record.update(serializable_result)
                                record['duplicate_of_event_id'] = event_id
                                duplicate_events += 1

                                # Проверяем, что frequency действительно увеличилась
                                updated_event = event_store.get_event_by_id(event_id)
                                if updated_event:
                                    print(f"📈 Подтверждено: частота события {event_id} теперь {updated_event.get('frequency', 1)}")
                            else:
                                print(f"❌ Не удалось обновить частоту события {event_id}")
                                # Если не удалось обновить, добавляем как новое
                                event_id = event_store.add_event(full_event_data, frequency=1, raw_source=record)
                                serializable_result = convert_to_serializable(result)
                                record.update(serializable_result)
                                record['event_id'] = event_id
                                successful_events += 1
                        else:
                            print(f"❌ Событие с индексом {duplicate_index} не найдено в хранилище")
                            # Если дубликат не найден, добавляем как новое событие
                            event_id = event_store.add_event(full_event_data, frequency=1, raw_source=record)
                            serializable_result = convert_to_serializable(result)
                            record.update(serializable_result)
                            record['event_id'] = event_id
                            successful_events += 1
                    else:
                        print("⚠️ Некорректный индекс дубликата, добавляем как новое событие")
                        event_id = event_store.add_event(full_event_data, frequency=1, raw_source=record)
                        serializable_result = convert_to_serializable(result)
                        record.update(serializable_result)
                        record['event_id'] = event_id
                        successful_events += 1

                else:
                    print(f"⚠️ Другой статус обработки: {processing_status}")
                    serializable_result = convert_to_serializable(result)
                    record.update(serializable_result)
                    empty_events += 1

            else:
                print("⚠️ Событие пустое или не найдено")
                if result:
                    serializable_result = convert_to_serializable(result)
                    record.update(serializable_result)
                record['processing_status'] = 'no_event'
                record['is_blacklist'] = is_blacklist
                empty_events += 1

        except Exception as e:
            record['scored_event'] = None
            record['processing_status'] = 'error'
            record['is_blacklist'] = False
            record['error_message'] = str(e)
            failed_events += 1
            print(f'❌ ОШИБКА ГЕНЕРАЦИИ СОБЫТИЙ: {e}')

    # В статистике обработки
    print(f"\n📊 СТАТИСТИКА ОБРАБОТКИ:")
    print(f"   ✅ Уникальных событий: {successful_events}")
    print(f"   🔁 Дубликатов: {duplicate_events}")
    print(f"   ❌ Ошибок: {failed_events}")
    print(f"   ⚠️ Пустые события: {empty_events}")
    print(f"   📈 Всего записей: {len(records)}")

    if successful_events > 0:
        success_rate = (successful_events / len(records)) * 100
        print(f"   🎯 Процент успеха: {success_rate:.1f}%")

    # СОХРАНЯЕМ УНИКАЛЬНЫЕ СОБЫТИЯ ДЛЯ ЭТОЙ ЗАДАЧИ
    try:
        # Получаем все уникальные события из хранилища
        unique_events = event_store.get_events()
        stats = event_store.get_stats()

        print(f"\n🎯 УНИКАЛЬНЫЕ СОБЫТИЯ ДЛЯ ЗАДАЧИ:")
        print(f"   📊 Уникальных событий: {stats['total_unique_events']}")
        print(f"   🔢 Всего вхождений: {stats['total_occurrences']}")
        print(f"   📋 Всего исходных записей: {stats['total_raw_sources']}")
        print(f"   📈 Средняя частота: {stats['avg_frequency']:.2f}")

        # Подготавливаем данные для сохранения
        events_to_save = []
        for i, event in enumerate(unique_events):
            print(f"🔍 Обрабатываю событие {i+1}/{len(unique_events)} (ID: {event.get('id')}, частота: {event.get('frequency', 1)})")

            try:
                # Рекурсивно преобразуем все данные в сериализуемый формат
                processed_event = convert_to_serializable(event)

                event_data = {
                    "id": processed_event.get('id'),
                    "event": processed_event.get('event', {}),
                    "score": processed_event.get('score', {}),
                    "frequency": processed_event.get('frequency', 1),
                    "region": processed_event.get('region', region),
                    "category": processed_event.get('category', category),
                    "raw_source": processed_event.get('raw_source', [])
                }
                events_to_save.append(event_data)
                print(f"✅ Событие {event.get('id')} подготовлено для сохранения (частота: {event.get('frequency', 1)})")

            except Exception as e:
                print(f"❌ Ошибка обработки события {event.get('id')}: {e}")
                # Добавляем пустое событие с ошибкой
                events_to_save.append({
                    "id": event.get('id'),
                    "event": {},
                    "score": {},
                    "frequency": event.get('frequency', 1),
                    "region": region,
                    "category": category,
                    "raw_source": [],
                    "error": f"Ошибка обработки: {str(e)}"
                })

        # Сортируем по частоте и оценке
        events_to_save = calculate_and_sort_events(events_to_save)

        # ОБРАБОТКА ТОП-5 СОБЫТИЙ ДЛЯ ФОРМИРОВАНИЯ SHORT_RAW_DATA
        # events_to_save = process_top_events_with_raw_data(events_to_save, top_n=5)

        print(f"💾 Сохраняю {len(events_to_save)} событий в {output_path}...")

        # Сохраняем уникальные события
        os.makedirs(output_dir_events, exist_ok=True)
        with open(output_path, 'w', encoding='utf-8') as f_out:
            json.dump(events_to_save, f_out, ensure_ascii=False, indent=2, cls=PydanticEncoder)

        print(f"✅ Уникальные события сохранены в: {output_path}")

        # Выводим события с frequency > 1 для проверки
        events_with_duplicates = [e for e in events_to_save if e.get('frequency', 1) > 1]
        if events_with_duplicates:
            print(f"\n🎯 СОБЫТИЯ С ДУБЛИКАТАМИ (frequency > 1):")
            for event in events_with_duplicates:
                print(f"   📍 ID: {event.get('id')}, Частота: {event.get('frequency')}, Событие: {event.get('event', {}).get('event', '')[:100]}...")
        else:
            print(f"\n⚠️ Нет событий с frequency > 1 (дубликаты не сохранились)")

    except Exception as e:
        print(f'❌ ОШИБКА СОХРАНЕНИЯ УНИКАЛЬНЫХ СОБЫТИЙ: {e}')
        import traceback
        traceback.print_exc()  # Выводим полную трассировку ошибки
````
